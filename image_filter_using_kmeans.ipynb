```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.cluster import KMeans
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread("/content/WhatsApp Image 2021-12-26 at 20.15.51.jpeg")
img=cv2.resize(img,(200,450))
# img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
# plt.figure(0,(6,10))
# # # plt.style.use('seaborn')
# plt.axis(False)
# plt.imshow(img)
# plt.show()
cv2_imshow(img)
```

```python
mm,nn,_=img.shape
```

```python
img=img.reshape(-1,3)
```

```python
kmeans=KMeans(n_clusters=10,random_state=5)
kmeans.fit(img)
```

```python
centers=kmeans.cluster_centers_
centers=np.array(centers,dtype=int)
print(centers)
```

```python
def distance(a,b):
  return np.sqrt(np.sum((a-b)**2))
```

```python
m,_=img.shape
new_img=np.zeros((m,3))
new_img.shape
```

```python
a=np.array([0,1])
b=np.array([3,5])
distance(a,b)
```

```python
for i in range(m):
  dis=[]
  for x in centers:
    c=distance(img[i],x)
    # print(c)
    dis.append(c)
  dis=np.array(dis)
  num=np.argmin(dis)
  new_img[i]=centers[num]
```

```python
new_img=new_img.reshape(mm,nn,3)
```
